## 🎯 Step 5 — Examine the target variable

The column we want to predict is `Revenue`:
- `True` → the session ended in a purchase
- `False` → the session did NOT end in a purchase

Let's see how balanced these two classes are. **Class balance matters** — if 99% of sessions are "no purchase", a model that always predicts "no purchase" would be 99% accurate but completely useless.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Count how many sessions ended with a purchase vs. not
counts = df['Revenue'].value_counts()
print("Class counts:")
print(counts)
print(f"\nProportion of buyers: {counts[True] / len(df):.2%}")
print(f"Proportion of non-buyers: {counts[False] / len(df):.2%}")

# Visualise it
fig, ax = plt.subplots(figsize=(6, 4))
sns.barplot(x=counts.index.astype(str), y=counts.values, ax=ax,
            palette=['#e74c3c', '#2ecc71'])
ax.set_title('Online Shoppers — Class distribution of Revenue')
ax.set_xlabel('Revenue (did the session end in a purchase?)')
ax.set_ylabel('Number of sessions')
plt.tight_layout()
plt.show()
```

## ✅ What we observed

After running the cells above you should have seen:

- **Shape**: 12,330 sessions × 18 columns
- **Mix of data types**: some numeric (like `Administrative_Duration`), some categorical (like `Month`, `VisitorType`), one Boolean target (`Revenue`)
- **Class imbalance**: only ~15% of sessions ended in a purchase

This imbalance is real and important. When we build our classifiers in Notebook 01, we'll need to either:
- Apply **SMOTE** (Synthetic Minority Over-sampling Technique) to balance the training data, or
- Use **class weights** to tell the model to pay more attention to the minority class
